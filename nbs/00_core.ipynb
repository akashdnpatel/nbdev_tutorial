```python
#| export
from nbdev.showdoc import *
import pandas as pd
```

```python
#| export
def prepare_dataframes():
    data1 = [10,20,3,40,10,20,30,40,50]
    df = pd.DataFrame(data1, columns=['Numbers'])
    
    data2 = [12,21,31,12,1,54,32,21,12]
    df1 = pd.DataFrame(data2, columns=['Numbers1'])
    return df, df1
```

```python
#| export
df, df1 = prepare_dataframes()
```

## Visualisation 1

```python
df.plot.hist()
```

```python
#| export
df2 = pd.concat([df,df1], axis=1)
df2
```

## Visualisation 2

```python
df2.plot.scatter(x='Numbers', y='Numbers1')
```
